```python
from twirp.context import Context
from twirp.exceptions import TwirpServerException
from twirp.errors import Errors
from google.protobuf.json_format import MessageToDict
# store the user profile


def mock_store_user_account(user_info, user_profile):
    # is_created = False
    user_id = None

    # create a new user account
    client = user_management_twirp.UserManagementServiceClient("http://localhost:60020")

    try:
        response = client.CreateUserAccount(
            ctx=Context(),
            request=user_management_pb2.UserAccount(
                user_name=user_info["user_name"],
                user_password="password",
                user_email=user_info["user_email"],
                user_phone=user_info["user_phone"],
                user_address=user_info["user_location"],
            ),
        )
        # print(response)
        user_id = response.user_id

    except TwirpServerException as e:
        print(e.code, e.message)
        if e.code == Errors.Internal:
            response = client.GetUserAccount(
                ctx=Context(),
                request=GetUserRequest(
                    user_name=user_info["user_name"],
                    user_password="password",
                    user_email=user_info["user_email"],
                    user_phone=user_info["user_phone"],
                ),
            )
            print("from get user account", response)

            open("snapshot/user_account.json", "w").write(
                json.dumps(
                    MessageToDict(response, including_default_value_fields=True),
                    indent=2,
                )
            )
            user_id = response.id
            # is_created = True
    print("user_id", user_id)

    return user_id


def mock_store_user_profile(user_info, user_id, user_profile):
    client = user_management_twirp.UserManagementServiceClient("http://localhost:60020", timeout=999999)
    user_profile_response = []
    try:
        response = client.ListUserProfile(
            ctx=Context(),
            request=GetUserRequest(
                user_id=user_id,
                user_name=user_info["user_name"],
            ),
        )
        # print(response)

        if response.profiles and len(response.profiles) > 0:
            user_profile_response = response.profiles
            return user_id, user_profile_response

    except TwirpServerException as e:
        print('list user profile fail, not found')
        print(e.code, e.message)
        pass

    print('then, create user profile')

    try:
        for job in user_profile:
            print('create user profile : ', job['title'])
            keywords_list = [
                PreferenceKeyword(
                    user_id=user_id,
                    keyword="UserProfile",
                    value=keyword,
                )
                for keyword in job["keywords"]
            ]
            keywords_list.append(
                PreferenceKeyword(
                    user_id=user_id,
                    keyword="industry",
                    value=job["industry"],
                )
            )

            request_set = user_management_pb2.UserProfile(
                user_id=user_id,
                title=job["title"],
                position=job["position"],
                description=job["description"],
                salary=job["salary"],
                location=job["location"],
                company=job["company"],
                type=UserProfileType.EMPLOYEE,
                keywords=keywords_list,
                start_date=job["start_date"],
                end_date=job["end_date"],
            )
            response = client.CreateUserProfile(
                ctx=Context(),
                request=request_set,
            )
            # user_profile_response.append(response)
            # print(response)
        response = client.ListUserProfile(
            ctx=Context(),
            request=GetUserRequest(
                user_id=user_id,
                user_name=user_info["user_name"],
            ),
        )
        # print(response)

        user_profile_response = response.profiles

    except TwirpServerException as e:
        print(e.code, e.message, e.meta, e.to_dict())

    return user_id, user_profile_response


def fetch_user_profile(user_info, user_id):
    client = user_management_twirp.UserManagementServiceClient("http://localhost:60020")
    user_profile_response = []
    try:
        response = client.ListUserProfile(
            ctx=Context(),
            request=GetUserRequest(
                user_id=user_id,
                user_name=user_info["user_name"],
            ),
        )
        print(response)

        if response.profiles and len(response.profiles) > 0:
            tmp = MessageToDict(response, including_default_value_fields=True)
            user_profile_response = response.profiles
            open("snapshot/user_profile.json", "w").write(json.dumps(tmp['profiles'], indent=2))

    except TwirpServerException as e:
        print('list user profile fail, not found')
        print(e.code, e.message)
        pass

    return user_profile_response
```

### back process for data preprocessing of the related job

this process is converting the related jobs, based on the user profile and account, to the model creation process.

> remark: the `related_job_data_preprocess` should implement in the prediction service.

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# from job_library import *
import job_library
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import os

script_base_dir = "beta_v1"

postcode_path = os.path.join(script_base_dir, "data", "au_postcodes.json")
postcode_data = job_library.load_other_data(postcode_path)


def job_location_data_preprocess(user_info, dataframe, dataframe_source):
    # location data
    print(dataframe.columns)
    user_location = user_info["user_location"]
    dataframe["locations"] = dataframe_source["locations"]
    dataframe[["suburb", "city", "region", "state"]] = dataframe_source[
        "locations"
    ].apply(lambda x: parse_location(x, job_library.STATE_MAPPING))
    dataframe["distance_km"] = job_library.calculate_distances(
        dataframe, user_location, postcode_data
    )
    dataframe = job_library.preprocess_zeros_nan(dataframe, user_location)
    dataframe.drop(columns=["locations"], inplace=True)
    return dataframe


def job_description_calculate_similarity(
    dataframe, dist_dataframe, resume_text, user_last_job_title, user_last_job_industry
):
    global model
    ## Create text similiarities features
    job_descriptions = dataframe["debugText"]

    # Using a pre-trained SBERT model

    dist_dataframe["descriptions_similarity_to_resume"] = (
        job_library.calculate_similarity(model, resume_text, job_descriptions)
    )

    # Reshape the data to fit the scaler
    descriptions_similarity_values = dist_dataframe[
        "descriptions_similarity_to_resume"
    ].values.reshape(-1, 1)

    scaler = MinMaxScaler(feature_range=(0, 1))

    # Fit the scaler to the data and transform
    dist_dataframe["descriptions_similarity_to_resume_norm"] = scaler.fit_transform(
        descriptions_similarity_values
    )

    ### Compare job title similarity between user's last job title and the job titles in the dataset
    dist_dataframe["job_title_similarity_to_last_job"] = (
        job_library.calculate_similarity(
            model, user_last_job_title, dataframe["postTitle"]
        )
    )

    dist_dataframe["job_industry_similarity_to_last_job"] = (
        job_library.calculate_similarity(
            model,
            user_last_job_industry,
            (
                dataframe["companyDetail.industry"].astype(str)
                + " "
                + dataframe["role"].astype(str)
            ),
        )
    )

    dist_dataframe["job_teamwork_score"] = job_library.calculate_similarity(
        model, job_library.COLLABORATION_TEAMWORK_TEXT, dataframe["debugText"]
    )

    dist_dataframe["job_growth_score"] = job_library.calculate_similarity(
        model, job_library.INNOVATION_GROWTH_TEXT, dataframe["debugText"]
    )

    dist_dataframe["job_respect_score"] = job_library.calculate_similarity(
        model, job_library.INCLUSIVITY_RESPECT_TEXT, dataframe["debugText"]
    )

    dist_dataframe["competitive_salary_score"] = job_library.calculate_similarity(
        model, job_library.COMPETITIVE_SALARY_TEXT, dataframe["debugText"]
    )

    dist_dataframe["benefit_score"] = calculate_similarity(
        model, job_library.BENEFIT_TEXT, dataframe["debugText"]
    )

    return dist_dataframe


def related_job_data_preprocess(user_info, related_jobs, last_job_profile):
    global postcode_data

    # prepare the data for the job
    df = pd.json_normalize(related_jobs)

    df = df.drop(
        columns=[
            "hittedKeywords",
            # "CompanyDetail",
            "companyDetail.lastUpdate",
            "companyDetail.linkedin",
            "companyDetail.locations",
            "companyDetail.referenceId",
            # "companyDetail.id",
            "companyDetail.url",
            # "postId",
            "postUrl",
            "expiringDate",
            # "id",
        ]
    )
    df_copy = job_library.feature_conversion(df)

    # location data
    df_copy = job_location_data_preprocess(user_info, df_copy, df)
    industry_keywords = [
        obj.value for obj in last_job_profile.keywords if obj.keyword == "industry"
    ]
    # job description data
    last_job_profile_industry = industry_keywords[0] if industry_keywords else None
    df_copy = job_description_calculate_similarity(
        df, df_copy,
        user_info["resume_text"],
        last_job_profile.title,
        last_job_profile_industry,
    )

    ## Calculate user preference score from the survey
    df_copy["user_preference_score"] = np.zeros(len(df_copy))

    df_copy["work_from_home_score"] = (
        df["debugText"]
        .str.contains("|".join(job_library.WFH_KEYWORDS), case=False, na=False)
        .astype(int)
    )

    # Hybrid keywords

    df_copy["hybrid_score"] = (
        df["debugText"]
        .str.contains("|".join(job_library.HYBRID_KEYWORDS), case=False, na=False)
        .astype(int)
    )

    # Create on_site_score
    df_copy["on_site_score"] = (
        (df_copy["work_from_home_score"] == 0) & (df_copy["hybrid_score"] == 0)
    ).astype(int)
    # Normalize 'pay_average' to the range 0 to 1
    df_copy["PayRangeMin_norm"] = (
        df_copy["PayRangeMin"] - df_copy["PayRangeMin"].min()
    ) / (df_copy["PayRangeMin"].max() - df_copy["PayRangeMin"].min())

    df_copy["PayRangeMax_norm"] = (
        df_copy["PayRangeMax"] - df_copy["PayRangeMax"].min()
    ) / (df_copy["PayRangeMax"].max() - df_copy["PayRangeMax"].min())
    # Alaternate: using percentile as the score
    # df_new['pay_percentile'] = df_new['pay_average'].rank(pct=True)

    df_copy["pay_average_norm"] = (
        df_copy["PayRangeMin_norm"] + df_copy["PayRangeMax_norm"]
    ) / 2

    df_copy["distance_norm"] = (
        df_copy["distance_km"] - df_copy["distance_km"].min()
    ) / (df_copy["distance_km"].max() - df_copy["distance_km"].min())

    # df_copy["CompanyDetail.GroupSize"] = df_0["CompanyDetail.GroupSize"]
    # df_copy["companyDetail.industry"] = df_0["companyDetail.industry"]
    df_copy["workType"] = df["workType"]
    # df_copy["CompanyDetail.HeadQuarters"] = df_0["CompanyDetail.HeadQuarters"]

    df_copy["headquarter_in_australia"] = df["companyDetail.headQuarters"].apply(
        lambda loc: int(
            job_library.is_location_in_australia(
                loc, postcode_data, job_library.STATE_MAPPING
            )
        )
    )

    return df_copy
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
## load other data
government_public_sector_path = os.path.join(
    script_base_dir, "data", "GovernmentOrPublicSector.json"
)
government_public_sector_data = load_other_data(government_public_sector_path)

nonprofit_organizations_path = os.path.join(
    script_base_dir, "data", "nonprofit_organizations.json"
)
nonprofit_organizations_data = load_other_data(nonprofit_organizations_path)


def process_survey_job_data(dataframe):
    ## Calculate User Preference score for each job
    dataframe["company_size_numeric"] = (
        dataframe["companyDetail.groupSize"].map(SIZE_MAPPING).astype(float)
    )
    mean_company_size = dataframe["company_size_numeric"].mean()
    dataframe["company_size_numeric"] = dataframe["company_size_numeric"].fillna(
        mean_company_size
    )

    dataframe["work_type_numeric"] = (
        dataframe["workType"].map(job_library.WORK_TYPE_MAPPING).astype(float)
    )
    mean_work_type = dataframe["work_type_numeric"].mean()
    dataframe["work_type_numeric"] = dataframe["work_type_numeric"].fillna(
        mean_work_type
    )
    dataframe["job_work_life_balance_score"] = 0

    return dataframe


def survey_process(dataframe, adjusted_features, feature_weights, user_preference_survey):
    adjusted_features = adjusted_features
    for keyword in user_preference_survey["keywords"]:
        if keyword["keyword"] == "job_title":
            dataframe, adjusted_features, feature_weights = survey_process_job_title(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "job_industry":
            dataframe, adjusted_features, feature_weights = survey_process_job_industry(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "company_size":
            dataframe, adjusted_features, feature_weights = survey_process_company_size(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "job_sector":
            dataframe, adjusted_features, feature_weights = survey_process_job_sector(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "company_culture":
            dataframe, adjusted_features, feature_weights = (
                survey_process_company_culture(
                    keyword["value"], dataframe, adjusted_features, feature_weights
                )
            )
        elif keyword["keyword"] == "work_model":
            dataframe, adjusted_features, feature_weights = survey_process_work_model(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "salary_expectations":
            dataframe, adjusted_features, feature_weights = (
                survey_process_salary_expectations(
                    keyword["value"], dataframe, adjusted_features, feature_weights
                )
            )
        elif keyword["keyword"] == "role_type":
            dataframe, adjusted_features, feature_weights = survey_process_role_type(
                keyword["value"], dataframe, adjusted_features, feature_weights
            )
        elif keyword["keyword"] == "company_scale":
            dataframe, adjusted_features, feature_weights = (
                survey_process_company_scale(
                    keyword["value"], dataframe, adjusted_features, feature_weights
                )
            )
        elif keyword["keyword"] == "location_distance":
            dataframe, adjusted_features, feature_weights = (
                survey_process_location_distance(
                    keyword["value"], dataframe, adjusted_features, feature_weights
                )
            )
    return dataframe, adjusted_features, feature_weights


# Q1: Job Title
def survey_process_job_title(key_value, df_new, adjusted_features, feature_weights):
    if key_value == "same_job_title":
        # df_new.loc[df_new['job_title_similarity_to_last_job'] >= 0.3 , 'user_preference_score'] += df_new['job_title_similarity_to_last_job']
        # new feature replace 'job_title_similarity_to_last_job'
        df_new["adjusted_job_title_similarity"] = df_new[
            "job_title_similarity_to_last_job"
        ]
        feature_weights["adjusted_job_title_similarity"] = 1.0
    elif key_value == "new_job_title":
        # df_new.loc[df_new['job_title_similarity_to_last_job'] < 0.3, 'user_preference_score'] += df_new['job_title_similarity_to_last_job'].max() - df_new['job_title_similarity_to_last_job']
        # new feature replace 'job_title_similarity_to_last_job'
        df_new["adjusted_job_title_similarity"] = (
            1 - df_new["job_title_similarity_to_last_job"]
        )
        feature_weights["adjusted_job_title_similarity"] = 0.8

    df_new["adjusted_job_title_similarity"] = (
        df_new["adjusted_job_title_similarity"]
        / df_new["adjusted_job_title_similarity"].max()
    )

    adjusted_features["adjusted_job_title_similarity"] = df_new[
        "adjusted_job_title_similarity"
    ]
    return df_new, adjusted_features, feature_weights


# Q2: Job Industry
def survey_process_job_industry(key_value, df_new, adjusted_features, feature_weights):
    if key_value == "last_industry":
        # df_new.loc[df_new['job_industry_similarity_to_last_job'] > df_new['job_industry_similarity_to_last_job'].quantile(0.7), 'user_preference_score'] += df_new['job_industry_similarity_to_last_job']
        df_new["adjusted_job_industry_similarity"] = df_new[
            "job_industry_similarity_to_last_job"
        ]
        feature_weights["adjusted_job_industry_similarity"] = 0.8

    elif key_value == "other_industry":
        # df_new.loc[df_new['job_industry_similarity_to_last_job'] < df_new['job_industry_similarity_to_last_job'].quantile(0.3), 'user_preference_score'] += 1 - df_new['job_industry_similarity_to_last_job']
        df_new["adjusted_job_industry_similarity"] = (
            1 - df_new["job_industry_similarity_to_last_job"]
        )
        feature_weights["adjusted_job_industry_similarity"] = 0.5

    adjusted_features["adjusted_job_industry_similarity"] = df_new[
        "adjusted_job_industry_similarity"
    ]
    # df_new['user_preference_score'] += df_new['adjusted_job_industry_similarity'] * feature_weights['adjusted_job_industry_similarity']
    return df_new, adjusted_features, feature_weights


# Q3: Company Size
def survey_process_company_size(key_value, df_new, adjusted_features, feature_weights):
    preferred_size = 0.7
    if key_value == "small_size":
        preferred_size = 1.5  # Average of SIZE_A and SIZE_B
    elif key_value == "medium_size":
        preferred_size = 3.5  # Average of SIZE_C and SIZE_D
    elif key_value == "large_size":
        preferred_size = 6.5  # Average of SIZE_E to SIZE_H

    max_size = df_new["company_size_numeric"].max()
    # convert difference to be similiarity by 1 - normalised value
    df_new["adjusted_company_size"] = 1 - (
        abs(df_new["company_size_numeric"] - preferred_size) / ((max_size - 1) if max_size > 1 else 1)
    )
    adjusted_features["adjusted_company_size"] = df_new["adjusted_company_size"]
    # df_new['user_preference_score'] += df_new['adjusted_company_size'] * feature_weights['adjusted_company_size']

    """
    if keyword['value'] == 'small_size':
        df_new.loc[(df_new['CompanyDetail.GroupSize'] == 'SIZE_A') | (df_new['CompanyDetail.GroupSize'] == 'SIZE_B'), 'user_preference_score'] += company_size_weight
    elif keyword['value'] == 'medium_size':
        df_new.loc[(df_new['CompanyDetail.GroupSize'] == 'SIZE_C') | (df_new['CompanyDetail.GroupSize'] == 'SIZE_D'), 'user_preference_score'] += company_size_weight
    elif keyword['value'] == 'large_size':
        df_new.loc[(df_new['CompanyDetail.GroupSize'].isin(['SIZE_E', 'SIZE_F', 'SIZE_G', 'SIZE_H'])), 'user_preference_score'] += company_size_weight 
    """
    return df_new, adjusted_features, feature_weights


# Q4: Job Sector
def survey_process_job_sector(key_value, df_new, adjusted_features, feature_weights):
    global government_public_sector_data, nonprofit_organizations_data
    if key_value == "government_or_public_sector":
        df_new["adjusted_job_sector"] = (
            df_new["companyDetail.industry"]
            .isin(government_public_sector_data["keywords"])
            .astype(int)
        )
    elif key_value == "nonprofit_organizations":
        df_new["adjusted_job_sector"] = (
            df_new["companyDetail.industry"]
            .isin(nonprofit_organizations_data["keywords"])
            .astype(int)
        )

    elif key_value == "private_sector":
        df_new["adjusted_job_sector"] = (
            ~df_new["companyDetail.industry"].isin(
                government_public_sector_data["keywords"]
                + nonprofit_organizations_data["keywords"]
            )
        ).astype(int)

    adjusted_features["adjusted_job_sector"] = df_new["adjusted_job_sector"]
    return df_new, adjusted_features, feature_weights


# Q5: Company Culture
def survey_process_company_culture(
    key_value, df_new, adjusted_features, feature_weights
):
    if key_value == "teamwork":
        df_new["adjusted_company_culture"] = df_new["job_teamwork_score"]
    elif key_value == "growth":
        df_new["adjusted_company_culture"] = df_new["job_growth_score"]
    elif key_value == "respect":
        df_new["adjusted_company_culture"] = df_new["job_respect_score"]
    elif key_value == "work_life_balance":
        df_new["adjusted_company_culture"] = df_new["job_work_life_balance_score"]

    # Normalize the culture scores
    df_new["adjusted_company_culture"] = (
        df_new["adjusted_company_culture"] / df_new["adjusted_company_culture"].max()
    )

    adjusted_features["adjusted_company_culture"] = df_new["adjusted_company_culture"]
    return df_new, adjusted_features, feature_weights


# Q6 Work Model
def survey_process_work_model(key_value, df_new, adjusted_features, feature_weights):
    if key_value == "remote":
        df_new["adjusted_work_model"] = df_new["work_from_home_score"]
    elif key_value == "hybrid":
        df_new["adjusted_work_model"] = df_new["hybrid_score"]
    elif key_value == "on_site":
        df_new["adjusted_work_model"] = df_new["on_site_score"]

    adjusted_features["adjusted_work_model"] = df_new["adjusted_work_model"]
    return df_new, adjusted_features, feature_weights


# Q7: Salary Expectations
def survey_process_salary_expectations(
    key_value, df_new, adjusted_features, feature_weights
):
    if key_value == "competitive_salary":
        df_new["adjusted_salary_expectation"] = df_new["competitive_salary_score"]
    elif key_value == "benefits":
        df_new["adjusted_salary_expectation"] = df_new["benefit_score"]
    elif key_value == "stable_income":
        # Assuming less competitive salary indicates stable income
        df_new["adjusted_salary_expectation"] = 1 - df_new["competitive_salary_score"]
    elif key_value == "work_life_balance":
        df_new["adjusted_salary_expectation"] = df_new["job_work_life_balance_score"]
        
    else:
        df_new["adjusted_salary_expectation"] = 0

    df_new["adjusted_salary_expectation"] = (
        df_new["adjusted_salary_expectation"]
        / df_new["adjusted_salary_expectation"].max()
    )
    adjusted_features["adjusted_salary_expectation"] = df_new["adjusted_salary_expectation"]
    return df_new, adjusted_features, feature_weights


# Q8: Role Type
def survey_process_role_type(key_value, df_new, adjusted_features, feature_weights):
    preferred_type = None
    if key_value == "fulltime":
        preferred_type = job_library.WORK_TYPE_MAPPING["Full time"]
    elif key_value == "contract":
        preferred_type = job_library.WORK_TYPE_MAPPING["Contract/Temp"]
    elif key_value == "parttime":
        preferred_type = (
            job_library.WORK_TYPE_MAPPING["Part time"]
            + job_library.WORK_TYPE_MAPPING["Casual/Vacation"]
        ) / 2

    max_type = max(WORK_TYPE_MAPPING.values())
    df_new['adjusted_role_type'] = 1 - (abs(df_new['work_type_numeric'] - preferred_type) / (max_type - 1))

    adjusted_features["adjusted_role_type"] = df_new["adjusted_role_type"]

    return df_new, adjusted_features, feature_weights


# Q9: Company Scale
def survey_process_company_scale( key_value, df_new, adjusted_features, feature_weights ):
    if key_value == 'foreign_based':
        df_new['adjusted_company_scale'] = 1 - df_new['headquarter_in_australia'].astype(int)
        feature_weights['adjusted_company_scale'] = 0.4  # Higher importance
        #df_new['user_preference_score'] += df_new['adjusted_company_scale'] * feature_weights['adjusted_company_scale']

    elif key_value == 'local':
        df_new['adjusted_company_scale'] = df_new['headquarter_in_australia'].astype(int)
        feature_weights['adjusted_company_scale'] = 0.4
        #df_new['user_preference_score'] += df_new['adjusted_company_scale'] * feature_weights['adjusted_company_scale']

    elif key_value == 'no_preference':
        df_new['adjusted_company_scale'] = df_new['headquarter_in_australia'].astype(int) 
        feature_weights['adjusted_company_scale'] = 0  # Lower importance
        # no adding to user preference score
        
    adjusted_features['adjusted_company_scale']  = df_new['adjusted_company_scale']   
    return df_new, adjusted_features, feature_weights


# Q10: Location Distance
def survey_process_location_distance(
    key_value, df_new, adjusted_features, feature_weights
):
    max_distance = 0
    if key_value == "30_minutes":
        max_distance = 25  # Approximate max distance for 30 minutes commute
        feature_weights['adjusted_distance_score'] = 1.0  # High importance
    elif key_value == "1_hour":
        max_distance = 50  # Approximate max distance for 1 hour commute
        feature_weights['adjusted_distance_score'] = 0.8
    elif key_value == "no_preference":
        max_distance = df_new['distance_km'].max()
        feature_weights['adjusted_distance_score'] = 0.4  # Lower importance

    df_new['adjusted_distance_score'] = df_new['distance_km'].apply(
        lambda x: max(0, (max_distance - x) / max_distance)
    )

    adjusted_features['adjusted_distance_score'] = df_new['adjusted_distance_score']
    return df_new, adjusted_features, feature_weights
```

```python
def process_user_preference_survey(dataframe, user_preference_survey):
    adjusted_features = {}
    # For example, include in your adjusted features
    adjusted_features["descriptions_similarity_to_resume_norm"] = dataframe[
        "descriptions_similarity_to_resume_norm"
    ]
    # Feature Weights default
    feature_weights = {
        'adjusted_job_title_similarity': 1.0,
        'adjusted_job_industry_similarity': 0.8,
        'adjusted_company_size': 0.7,
        'adjusted_job_sector': 0.7,
        'adjusted_company_culture': 0.6,
        'adjusted_work_model': 0.9,
        'adjusted_salary_expectation': 0.7,
        'adjusted_role_type': 0.8,
        'adjusted_distance_score': 1.0, 
        'descriptions_similarity_to_resume_norm': 1.0
    }

    dataframe, adjusted_features, feature_weights = survey_process(dataframe, adjusted_features,feature_weights, user_preference_survey)


    # Normalize the weights
    total_weight = sum(feature_weights.values())
    feature_weights = {k: v / total_weight for k, v in feature_weights.items()}
    # feature_weights = {k: v for k, v in feature_weights.items()}


    dataframe["user_preference_score"] = sum(
        dataframe[feature] * feature_weights[feature] for feature in adjusted_features
    )

    ### K-mean clustering

    # List of features and their types
    features_info = {
        "adjusted_job_title_similarity": "continuous",
        "adjusted_job_industry_similarity": "continuous",
        "adjusted_company_size": "continuous",
        "adjusted_job_sector": "categorical",
        "adjusted_company_culture": "continuous",
        "adjusted_work_model": "categorical",
        "adjusted_salary_expectation": "continuous",
        "adjusted_role_type": "categorical",
        "adjusted_distance_score": "continuous",
        "descriptions_similarity_to_resume_norm": "continuous",
        "pay_average_norm": "continuous",
    }

    # Initialize variables
    pairs_list = []
    pair_id = 1
    max_pairs = 15  # Number of pairs you want to select

    # K-Means Clustering
    num_clusters = 5  # Adjust based on your data
    scores = dataframe["user_preference_score"].values.reshape(-1, 1)
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    dataframe["score_cluster"] = kmeans.fit_predict(scores)

    # Initialize thresholds and criteria
    similarity_threshold = 0.95
    difference_threshold = 0.05
    min_similarities = 4
    min_differences = 3

    # Initialize variables for adjustments
    adjustment_iterations = 0
    max_adjustments = 5  # Maximum number of adjustments to prevent infinite loops

    # Loop to adjust criteria if not enough pairs are found
    while len(pairs_list) < max_pairs and adjustment_iterations < max_adjustments:
        # Reset pairs_list and job_appearance_counts for a fresh start
        pairs_list = []
        job_appearance_counts = {}
        pair_id = 1

        # Iterate over each cluster
        for cluster_label in dataframe["score_cluster"].unique():
            # Get jobs in the current cluster
            cluster_jobs = dataframe[dataframe["score_cluster"] == cluster_label]

            # Ensure there are at least 2 jobs to form pairs
            if len(cluster_jobs) < 2:
                continue  # Skip this cluster

            # Get all possible pairs within the cluster
            job_indices = cluster_jobs.index.tolist()
            cluster_pairs = list(combinations(job_indices, 2))

            # Shuffle pairs to introduce randomness
            np.random.shuffle(cluster_pairs)

            # Iterate over the pairs
            for job_idx_A, job_idx_B in cluster_pairs:
                job_A = cluster_jobs.loc[job_idx_A]
                job_B = cluster_jobs.loc[job_idx_B]

                # Get the 'PostId' as job IDs
                job_id_A = job_A["postId"]
                job_id_B = job_B["postId"]

                # Check if either job has already appeared in 2 pairs
                count_A = job_appearance_counts.get(job_id_A, 0)
                count_B = job_appearance_counts.get(job_id_B, 0)

                if count_A >= 2 or count_B >= 2:
                    continue  # Skip this pair

                # Compute feature similarities and overall similarity
                feature_similarities, avg_similarity = compute_similarity(
                    job_A, job_B, features_info
                )

                # Determine similarities and differences based on thresholds
                similarities = []
                differences = []
                for feature, sim_score in feature_similarities.items():
                    if sim_score >= similarity_threshold:
                        similarities.append(feature)
                    elif sim_score <= difference_threshold:
                        differences.append(feature)

                # Criteria: At least min_similarities similarities and min_differences differences
                if (
                    len(similarities) >= min_similarities
                    and len(differences) >= min_differences
                ):
                    pair_info = {
                        "pair_id": int(pair_id),
                        "job_A": {
                            "job_id": convert_value(job_id_A),
                            "user_preference_score": float(
                                convert_value(job_A["user_preference_score"])
                            ),
                            "features": {
                                feature: convert_value(job_A[feature])
                                for feature in features_info.keys()
                            },
                        },
                        "job_B": {
                            "job_id": convert_value(job_id_B),
                            "user_preference_score": float(
                                convert_value(job_B["user_preference_score"])
                            ),
                            "features": {
                                feature: convert_value(job_B[feature])
                                for feature in features_info.keys()
                            },
                        },
                        "similarities": similarities,
                        "differences": differences,
                        "overall_similarity": float(avg_similarity),
                    }
                    pairs_list.append(pair_info)
                    pair_id += 1

                    # Update job appearance counts
                    job_appearance_counts[job_id_A] = count_A + 1
                    job_appearance_counts[job_id_B] = count_B + 1

                    # Stop if we've collected enough pairs
                    if len(pairs_list) >= max_pairs:
                        break

            # Stop if we've collected enough pairs
            if len(pairs_list) >= max_pairs:
                break

        # Check if enough pairs were found
        if len(pairs_list) < max_pairs:
            # Adjust thresholds and criteria
            similarity_threshold -= 0.05  # Decrease similarity threshold
            difference_threshold += 0.05  # Increase difference threshold
            min_similarities = max(2, min_similarities - 1)
            min_differences = max(2, min_differences - 1)
            adjustment_iterations += 1
            # print(f"Adjusting thresholds: similarity >= {similarity_threshold}, differences <= {difference_threshold}")
        else:
            break
    return dataframe, pairs_list
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.model_selection import cross_val_score, LeaveOneOut
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import shap

# Train function (XGBOost)
def train_process_job_preference_ranking_model_XGBboost(df_new_for_XGboost, job2pick_choices):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the XGBoost model
    model = xgb.XGBClassifier(
    n_estimators=50,       # Number of boosting rounds
    learning_rate=0.1,      # Step size shrinkage (controls the impact of each tree)
    max_depth=4,            # Maximum tree depth for base learners
    reg_lambda=1,           # L2 regularization term
    reg_alpha=0,            # L1 regularization term
    random_state=42         # Seed for reproducibility
    )
    model.fit(X, y)

    return model, job_features, X, y




def train_process_job_preference_ranking_model_Decision_Tree(df_new_for_XGboost, job2pick_choices):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the Decision Tree Classifier
    model = DecisionTreeClassifier(
        criterion='gini',      # You can also try 'entropy' for information gain
        splitter='best',       # Choose 'best' to split on the most important feature, or 'random' for random splitting
        max_depth=6,           # Control the maximum depth to prevent overfitting
        min_samples_split=2,   # Minimum number of samples required to split an internal node
        min_samples_leaf=1,    # Minimum number of samples required to be at a leaf node
        max_features=None,     # Consider all features for best split; set to 'auto', 'sqrt', or 'log2' for tuning
        class_weight=None,     # Balance class weights automatically if there's class imbalance ('balanced' or custom dict)
        random_state=42,       # Seed for reproducibility
        max_leaf_nodes=None,   # Limit the maximum number of leaf nodes (None means unlimited)
    )
    model.fit(X, y)

    return model, job_features, X, y




def train_process_job_preference_ranking_model_Random_Forest(df_new_for_XGboost, job2pick_choices):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the RandomForestClassifier with predefined hyperparameters
    model = RandomForestClassifier(
        n_estimators=50,      # Number of trees
        max_depth=6,           # Limit depth of trees
        min_samples_split=2,   # Minimum samples to split a node
        min_samples_leaf=2,    # Minimum samples per leaf node
        random_state=42, 
        max_features='sqrt',
        bootstrap=True # Random state for reproducibility
    )
    model.fit(X, y)

    return model, job_features, X, y







def train_process_job_preference_ranking_model_Linear_Regression(df_new_for_XGboost, job2pick_choices):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the Linear Regression model with L2 regularization
    model = LogisticRegression(
        penalty='l2',
        C=1.0,
        solver='liblinear',
        random_state=42
    )
    model.fit(X, y)

    return model, job_features, X, y




# Prediction function
def predict_job_preference_ranking_model(model, new_jobs):
    # Generate all possible pairs of new jobs
    new_job_ids = new_jobs.index.values
    new_job_pairs = [
        (new_job_ids[i], new_job_ids[j])
        for i in range(len(new_job_ids))
        for j in range(i + 1, len(new_job_ids))
    ]

    # Create a DataFrame to hold scores
    job_scores = pd.Series(0, index=new_job_ids, dtype=float)

    # Calculate feature differences and predict preferences
    for job_A, job_B in new_job_pairs:
        # Create the feature difference using the function
        row = {"job_a_id": job_A, "job_b_id": job_B}
        feature_diff_array = create_feature_diff(
            row, new_jobs.drop(columns=["postId"])
        ).reshape(1, -1)

        # Predict preference
        prob_A_preferred = model.predict_proba(feature_diff_array)[0, 1]

        # Update scores
        job_scores[job_A] += prob_A_preferred
        job_scores[job_B] += 1 - prob_A_preferred

    # Sort the job scores in ascending order
    ranked_jobs_descending = job_scores.sort_values(ascending=False)

    # Retrieve the corresponding rows from the new_jobs DataFrame
    ranked_jobs_df_descending = new_jobs.loc[ranked_jobs_descending.index]

    # Add the job scores as a new column
    ranked_jobs_df_descending["job_score"] = ranked_jobs_descending.values

    return ranked_jobs_df_descending
```

```python
import os
import job_library
import xgboost as xgb
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df_new_for_XGboost = pd.read_parquet("snapshot/job_preference_df_for_xgboost.parquet")


# Update the function to handle numerical and categorical data differently
def create_feature_diff(row, job_features):
    tmp = job_features.copy()

    job_A_df = tmp.loc[row["job_a_id"]].dropna()
    job_A = job_A_df.iloc[0]

    job_B_df = tmp.loc[row["job_b_id"]].dropna()
    job_B = job_B_df.iloc[0]

    if type(job_A) != pd.Series:
        job_A = job_A_df

    if type(job_B) != pd.Series:
        job_B = job_B_df
    

    # num_diff = job_A[numerical_columns] - job_B[numerical_columns]
    num_diff = job_A - job_B

    # Calculate differences for categorical columns (1 if different, 0 if the same), but all categorical features are already encoded numerically
    # cat_diff = (job_A[categorical_columns] != job_B[categorical_columns]).astype(int)
    # Concatenate numerical and categorical differences
    # feature_diff = pd.concat([num_diff, cat_diff])

    feature_diff = num_diff
    # print("feature_diff" , feature_diff)

    return feature_diff.values.flatten()


# Train function
def train_process_job_preference_ranking_model_XGBoost(df_new_for_XGboost, job2pick_choices):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.copy()
    job_features = job_features.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the XGBoost model
    model = xgb.XGBClassifier()
    model.fit(X, y)

    return model, job_features


def train_process_job_preference_ranking_model_Decision_Tree( df_new_for_XGboost, job2pick_choices ):
    # Ensure 'postId' is in the DataFrame
    if "postId" not in df_new_for_XGboost.columns:
        print("Error: 'postId' column not found in df_new_for_XGboost.")
        return None, None

    # Set 'postId' as the index for easy access
    job_features = df_new_for_XGboost.set_index("postId")

    # Convert index to string if necessary
    job_features.index = job_features.index.astype(str)

    # Create a DataFrame from the choices
    comparisons = pd.DataFrame(
        job2pick_choices, columns=["pair_id", "user_choice", "job_a_id", "job_b_id"]
    )

    # Ensure job IDs are of the correct type
    comparisons["job_a_id"] = comparisons["job_a_id"].astype(str)
    comparisons["job_b_id"] = comparisons["job_b_id"].astype(str)

    # Create feature difference data (X)
    diffs = [
        create_feature_diff(row, job_features) for _, row in comparisons.iterrows()
    ]
    X = np.array(diffs)

    # Create the target variable y
    y = comparisons["user_choice"].values.astype(int)

    # Train the Decision Tree Classifier
    model = DecisionTreeClassifier(
        criterion="gini",  # You can also try 'entropy' for information gain
        splitter="best",  # Choose 'best' to split on the most important feature, or 'random' for random splitting
        max_depth=3,  # Control the maximum depth to prevent overfitting
        min_samples_split=2,  # Minimum number of samples required to split an internal node
        min_samples_leaf=1,  # Minimum number of samples required to be at a leaf node
        max_features=None,  # Consider all features for best split; set to 'auto', 'sqrt', or 'log2' for tuning
        class_weight=None,  # Balance class weights automatically if there's class imbalance ('balanced' or custom dict)
        random_state=42,  # Seed for reproducibility
        max_leaf_nodes=None,  # Limit the maximum number of leaf nodes (None means unlimited)
    )
    model.fit(X, y)

    return model, job_features


# Prediction function
def predict_job_preference_ranking_model(model, new_jobs):
    # Generate all possible pairs of new jobs
    new_job_ids = new_jobs.index.values
    new_job_pairs = [
        (new_job_ids[i], new_job_ids[j])
        for i in range(len(new_job_ids))
        for j in range(i + 1, len(new_job_ids))
    ]

    # Create a DataFrame to hold scores
    job_scores = pd.Series(0, index=new_job_ids, dtype=float)

    # Calculate feature differences and predict preferences
    for job_A, job_B in new_job_pairs:
        # Create the feature difference using the function
        row = {"job_a_id": job_A, "job_b_id": job_B}
        feature_diff_array = create_feature_diff(
            row, new_jobs.drop(columns=["postId"])
        ).reshape(1, -1)

        # Predict preference
        prob_A_preferred = model.predict_proba(feature_diff_array)[0, 1]

        # Update scores
        job_scores[job_A] += prob_A_preferred
        job_scores[job_B] += 1 - prob_A_preferred

    # Sort the job scores in ascending order
    ranked_jobs_descending = job_scores.sort_values(ascending=False)

    # Retrieve the corresponding rows from the new_jobs DataFrame
    ranked_jobs_df_descending = new_jobs.loc[ranked_jobs_descending.index]

    # Add the job scores as a new column
    ranked_jobs_df_descending["job_score"] = ranked_jobs_descending.values

    return ranked_jobs_df_descending
```

```python
def merge_ranked_jobs_with_related_jobs(job_rank_df, related_jobs):
    job_rank_df = job_rank_df.copy()
    exported_job = []
    for _, row in job_rank_df.iterrows():
        
        filtered = [job for job in related_jobs if str(job["postId"]) == str(row["postId"])]

        if len(filtered) > 0:
            related_job = filtered[0]
            related_job["score"] = row["job_score"]
            exported_job.append(related_job)
    return exported_job
```
